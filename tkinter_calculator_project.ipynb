```python
# tkinter calculator (does not run in google collab but in vscode)

import tkinter as tk
from cmath import sqrt

text_ =  ''

def add_text_to_field(text):
    global text_
    text_ = text_ + str(text)
    field.delete('1.0', 'end')
    field.insert('1.0', text_)

def calculus():
    global text_
    result = str(eval(text_))
    field.delete('1.0', 'end')
    field.insert('1.0', result)

def square_root():
    global text_
    sqrt_result = str(sqrt(float(text_)))
    field.delete('1.0', 'end')
    field.insert('1.0', sqrt_result)

def clear():
    global text_
    text_ = ''
    field.delete('1.0', 'end')


screen_window = tk.Tk()
screen_window.geometry('400x400')
field = tk.Text(screen_window, height = 2, width = 30, font = ('Arial', 18))
field.grid(row = 1, column = 1, columnspan = 4)

# number buttons
button_one = tk.Button(screen_window, text = '1', command = lambda: add_text_to_field(1), width = 10, font = ('Arial', 12))
button_one.grid(row = 4, column = 1)

button_two = tk.Button(screen_window, text = '2', command = lambda: add_text_to_field(2), width = 10, font = ('Arial', 12))
button_two.grid(row = 4, column = 2)

button_three = tk.Button(screen_window, text = '3', command = lambda: add_text_to_field(3), width = 10, font = ('Arial', 12))
button_three.grid(row = 4, column = 3)

button_four = tk.Button(screen_window, text = '4', command = lambda: add_text_to_field(4), width = 10, font = ('Arial', 12))
button_four.grid(row = 3, column = 1)

button_five = tk.Button(screen_window, text = '5', command = lambda: add_text_to_field(5), width = 10, font = ('Arial', 12))
button_five.grid(row = 3, column = 2)

button_six = tk.Button(screen_window, text = '6', command = lambda: add_text_to_field(6), width = 10, font = ('Arial', 12))
button_six.grid(row = 3, column = 3)

button_seven = tk.Button(screen_window, text = '7', command = lambda: add_text_to_field(7), width = 10, font = ('Arial', 12))
button_seven.grid(row = 2, column = 1)

button_eight = tk.Button(screen_window, text = '8', command = lambda: add_text_to_field(8), width = 10, font = ('Arial', 12))
button_eight.grid(row = 2, column = 2)

button_nine = tk.Button(screen_window, text = '9', command = lambda: add_text_to_field(9), width = 10, font = ('Arial', 12))
button_nine.grid(row = 2, column = 3)

button_zero = tk.Button(screen_window, text = '0', command = lambda: add_text_to_field(0), width = 10, font = ('Arial', 12))
button_zero.grid(row = 5, column = 1)

# calculus buttons

button_decimal = tk.Button(screen_window, text = '.', command = lambda: add_text_to_field('.'), width = 10, font = ('Arial', 12))
button_decimal.grid(row = 5, column = 2)

button_clear = tk.Button(screen_window, text = 'Clear', command = lambda: clear(), width = 10, font = ('Arial', 12))
button_clear.grid(row = 5, column = 3)

button_add = tk.Button(screen_window, text = 'Add', command = lambda: add_text_to_field('+'), width = 10, font = ('Arial', 12))
button_add.grid(row = 2, column = 4)

button_minus = tk.Button(screen_window, text = 'Subtract', command = lambda: add_text_to_field('-'), width = 10, font = ('Arial', 12))
button_minus.grid(row = 3, column = 4)

button_multiply = tk.Button(screen_window, text = 'Multiply', command = lambda: add_text_to_field('*'), width = 10, font = ('Arial', 12))
button_multiply.grid(row = 4, column = 4)

button_divide = tk.Button(screen_window, text = 'Divide', command = lambda: add_text_to_field('/'), width = 10, font = ('Arial', 12))
button_divide.grid(row = 5, column = 4)

button_parenthesis_open =  tk.Button(screen_window, text = '(', command = lambda: add_text_to_field('('), width = 10, font = ('Arial', 12))
button_parenthesis_open.grid(row = 6, column = 1)

button_parenthesis_close =  tk.Button(screen_window, text = ')', command = lambda: add_text_to_field(')'), width = 10, font = ('Arial', 12))
button_parenthesis_close.grid(row = 6, column = 2)

button_squareroot =  tk.Button(screen_window, text = 'Square root', command = lambda: square_root(), width = 10, font = ('Arial', 12))
button_squareroot.grid(row = 6, column = 3)

button_answer =  tk.Button(screen_window, text = 'Answer', command = lambda: calculus(), width = 10, font = ('Arial', 12))
button_answer.grid(row = 6, column = 4)

screen_window.mainloop()
```
